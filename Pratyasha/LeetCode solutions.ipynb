```python
from typing import List

class Solution:
    def runningSum(self, nums: List[int]) -> List[int]:
        sum = 0
        for i in range(len(nums)):
            sum = sum + nums[i]
            nums[i] = sum
        return nums   
        
```

```python
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        max_candie = max(candies)
        tf_final =[]
        for i in range(len(candies)):
            if candies[i]+extraCandies >= max_candie:
                tf_final.append('true')
            else:
                tf_final.append('false')
        return tf_final
```

## 1672. Richest Customer Wealth

```python
class Solution:
    def maximumWealth(self, accounts: List[List[int]]) -> int:
        Sum = []
        for i in range(len(accounts)):
            Sum.append(sum(accounts[i]))
        return max(Sum)
            
```

## 1470. Shuffle the Array

```python
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        Shuffle=[]
        for i in range(n):
            Shuffle.append(nums[i])
            Shuffle.append(nums[n+i])
        return Shuffle
```

```python
Solution().shuffle([1,2,3,4,4,3,2,1],4)
```

## 1512. Number of Good Pairs

```python
class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        goodPairs=0
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                    if(nums[i]==nums[j]):
                        goodPairs+=1
        return goodPairs
                            
```

```python
Solution().numIdenticalPairs([1,2,3,1,1,3])
```

## 771. Jewels and Stones

```python
class Solution:
    def numJewelsInStones(self, jewels: str, stones: str) -> int:
        jw = list(jewels)
        total_jewels = 0
        for j in jw:
            for i in range(len(stones)):
                if j==stones[i]:
                    total_jewels+=1
        return total_jewels
                    
```

```python
Solution().numJewelsInStones("aA","aAAbbbb")
```

## 1773. Count Items Matching a Rule

```python
class Solution:
    def countMatches(self, items: List[List[str]], ruleKey: str, ruleValue: str) -> int:
        counter = 0
        if ruleKey=='type':
            for i in items:
                if i[0] == ruleValue:
                    counter+=1
        elif ruleKey=='color':
            for i in items:
                if i[1] == ruleValue:
                    counter+=1
        elif ruleKey=='name':
            for i in items:
                if i[2] == ruleValue:
                    counter+=1
        return counter
```

## 1603. Design Parking System

```python
class ParkingSystem:

    def __init__(self, big: int, medium: int, small: int):
        self.b = big
        self.m=medium
        self.s=small

    def addCar(self, carType: int) -> bool:
        if carType == 1 and self.b:
            self.b-=1
            return True
        elif carType == 2 and self.m:
            self.m-=1
            return True
        elif carType == 3 and self.s:
            self.s-=1
            return True
        else:return False

```

```python
parkingSystem = ParkingSystem(1, 1, 0)
```
